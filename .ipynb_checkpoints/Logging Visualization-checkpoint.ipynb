```python
import numpy as np 
import pandas as pd 
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
df = pd.read_fwf('output.log') 
print(3 %3)
```

```python
foo = df['Unnamed: 2']
i = []
j = []
k = []
for index,row in enumerate(foo):
    if index % 3 == 0:
        i.append(row)
    elif index % 3 == 1:
        j.append(row)
    elif index % 3 == 2:
        k.append(row)

a = len(i)
b = len(j)
c = len(k)

```

```python
plt.title('Lottery Scheduler')
y_axis = [i for i in range(61)]

plt.plot( y_axis, i) # 30
plt.plot(y_axis, j) # 20
plt.plot(y_axis, k) # 10
plt.xlabel('Time in ticks that have passed since measurement started (divided by 100)')
plt.ylabel('Total time in ticks that each process has spent scheduled (across all time)')
plt.legend(["10 tickets", "20 tickets", "30 tickets"], loc=0, frameon= True)
plt.savefig("graph.jpg")
plt.savefig("graph.png")
plt.savefig("graph.pdf")
plt.show()
```
